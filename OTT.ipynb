## No. of movies/series categorized by age restriction before 2000 and after 2000?

```python
year_group = df.Year.unique()
year_group
```

```python
year_group.min()
```

```python
year_group.max()
```

```python
Before2000=df[(df.Year <= 2000)]
Before2000
```

```python
Before2000.count()
```

```python
After2000 = df[(df.Year > 2000)] # dont do => 2000 because we already did '=' in before200
After2000
```

```python
After2000.count()
```

```python
MovieBefore2000 = Before2000.groupby("Age").Title.count()
MovieBefore2000
```

```python
plt.figure(figsize=(14,6))
plt.plot(MovieBefore2000,'g--')
plt.title("Age restriction movies before 2000")
plt.xlabel("Age-->")
plt.ylabel("No. of Movies/Series-->")
plt.show()
```

```python
Moviesafter2000=After2000.groupby("Age").Title.count()
Moviesafter2000
```

```python
plt.figure(figsize=(14,6))
plt.plot(Moviesafter2000,'b--')
plt.title("Age restriction movies after 2000")
plt.xlabel("Age-->")
plt.ylabel("No. of Movies/Series-->")
plt.show()
```

## Comparing both

```python
plt.figure(figsize=(14,6))
plt.plot(MovieBefore2000,'r--')
plt.plot(Moviesafter2000,'g')
plt.title("Comparing Age Categories of Movies/Series Before and After 2000")
plt.xlabel("Age Categories")
plt.ylabel("No. of Movies/Series")
plt.show()
```

## Using Percentage of movies in graph

```python
percentageBefore2000=MovieBefore2000 / sum(MovieBefore2000) #Convert in %
percentageBefore2000
```

```python
percentageAfter2000= Moviesafter2000 / sum(Moviesafter2000)
percentageAfter2000
```

```python
plt.figure(figsize=(14,6))
plt.plot(percentageBefore2000,'y--')
plt.plot(percentageAfter2000,'g')
plt.title("Comparing Age Categories by Movies/Series Before and After 2000")
plt.xlabel("Age Categories")
plt.ylabel("% of Movies/Series")
plt.legend(["Before 2000","After2000"])
plt.show()
```
